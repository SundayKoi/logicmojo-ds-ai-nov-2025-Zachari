## Problem: 2

Generate an array of length $3n$ , filled with the cyclic pattern 1, 2, 3. Where $n$ is a natural number.

```python
def generate_cyclic_array(n):
    return np.tile([1, 2, 3], n)

print(generate_cyclic_array(5))  # [1 2 3 1 2 3 1 2 3 1 2 3 1 2 3]
```

## Problem: 4

Create a 10 x 10 arrays of zeros and then "frame" it with a border of ones.

```python
# Create 10x10 array of zeros
array = np.zeros((10, 10), dtype=int)

# Frame it with ones
array[0, :] = 1    # Top border
array[-1, :] = 1   # Bottom border
array[:, 0] = 1    # Left border
array[:, -1] = 1   # Right border

print(array)
```

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
checkerboard = np.zeros((8, 8), dtype=int)

# Set alternating positions to 1
checkerboard[::2, 1::2] = 1  # Even rows, odd columns
checkerboard[1::2, ::2] = 1  # Odd rows, even columns

print(checkerboard)

```

## Problem: 6

Create an array $x$ with 100 points in the interval $[-1,1]$ (the points must be uniformly spaced). Then create another array $y$.

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
def create_array(n):
    array = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            array[i, j] = i + j
    return array

# Test it
print(create_array(4))

```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
x = np.linspace(0, 1, 1000)
cos_x = np.cos(x)
sin_x = np.sin(x)
result = np.column_stack((cos_x, sin_x))

```

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python

from numpy import random

x = random.rand(3, 5)
total = x.sum()
row_sums = x.sum(axis=1)
column_sums = x.sum(axis=0)

print(x)
print(total)
print(row_sums)
print(column_sums)
```
